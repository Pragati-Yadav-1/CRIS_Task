```python
import geopandas as gpd
import networkx as nx
import matplotlib.pyplot as plt
from shapely.geometry import LineString, Point

# Step 1: Load the shapefile into a GeoDataFrame
lines_gdf = gpd.read_file('lines.shp')

# Step 2: Extract unique points (nodes) from LineStrings
points = []
for line in lines_gdf.geometry:
    points.extend(list(line.coords))
unique_points = list(set(points))  # Remove duplicates

# Step 3: Create a graph
G = nx.Graph()

# Add nodes (points)
for i, point in enumerate(unique_points):
    G.add_node(i, pos=point)  # Node index i with position point

# Add edges (LineStrings)
for line in lines_gdf.geometry:
    points = list(line.coords)
    for i in range(len(points) - 1):
        start_point = points[i]
        end_point = points[i + 1]
        start_node = unique_points.index(start_point)
        end_node = unique_points.index(end_point)
        G.add_edge(start_node, end_node, weight=line.length)  # Use LineString length as weight
        #The line.length attribute in the context of a LineString object from the Shapely library gives the actual 
        # Euclidean distance, not the number of points.

# Step 4: Visualize the graph
pos = nx.get_node_attributes(G, 'pos')
nx.draw(G, pos, with_labels=True, node_color='red', node_size=50, edge_color='blue', font_size=8)
plt.title('Graph Representation of Points and LineStrings')
plt.show()

# Step 5: Find shortest path between two nodes
start_node = 0  # can replace with actual node index based on your data
end_node = 3    # can replace with actual node index based on your data

try:
    shortest_path = nx.shortest_path(G, source=start_node, target=end_node, weight='weight')
    total_distance = nx.shortest_path_length(G, source=start_node, target=end_node, weight='weight')

    print(f"Shortest path nodes: {shortest_path}")
    print(f"Total distance: {total_distance:.2f} units")

except nx.NetworkXNoPath:
    print(f"No path found between node {start_node} and node {end_node}.")
except nx.NodeNotFound as e:
    print(f"Node not found in the graph: {e}")
except Exception as e:
    print(f"Error: {e}")
```

## The line.length attribute in the context of a LineString object from the Shapely library gives the actual Euclidean distance, not the number of points.
## LineString in Shapely: A LineString is a sequence of points connected by straight, non-intersecting line segments.
## Euclidean Distance: The length attribute calculates the total length of the LineString by summing the Euclidean distances between consecutive points in the sequence.
